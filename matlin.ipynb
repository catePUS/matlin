```python
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
f = 0.5
G = f * A
G

```

```python
G = np.array(G)
x = G[:,0]
y = G[:,1]

plt.plot(x, y)
plt.ylim([-100, 100])
plt.xlim([-100, 100])
```

```python
angle = np.radians(130)
Matrix = np.array([[np.cos(angle), -np.sin(angle), 0], [np.sin(angle), np.cos(angle),  0], [0,  0,  1]])


turnmatr=np.dot(A,Matrix)
turnmatr = np.array(turnmatr)
x = turnmatr[:,0]
y = turnmatr[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python

newmatrix = np.array([[0.5, 0, 0], 
                   [0, 1.2,  0], 
                   [200,  300,  1]])


R=np.dot(A,newmatrix)
R = np.array(R)
x = R[:,0]
y = R[:,1]

plt.plot(x, y)
plt.ylim([-500, 500])
plt.xlim([-500, 500])
```
